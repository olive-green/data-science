```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/train.csv")
```

```python
pd.set_option("display.max_columns",None)
pd.set_option("display.max_rows",None)
```

```python
df.shape
```

```python
df.info()
```

```python
df.isna().sum()
```

#Now removing columns containing null values>20

```python
null_values_per=df.isna().sum()/df.shape[0]*100
null_values_per
```

```python
null_values_columns=null_values_per[null_values_per > 20].keys()
null_values_columns
```

```python
# now we drop these columns
df1=df.drop(columns=null_values_columns)
df1.shape
```

```python
df1.isna().sum()/df1.shape[0]*100
```

#Now we fills these null values by the use of mean,median

```python
# first we fill numeric data
df1numeric_data=df1.select_dtypes(include=['int64','float64'])
df1numeric_data.head()
```

```python
# now we plot these variables to see null values
plt.figure(figsize=(16,9))
sns.heatmap(df1numeric_data.isna())
```

```python
# now to get rows of all null values
df1numeric_data[df1numeric_data.isna().any(axis=1)]
```

#null values columns name

```python
missing_null_var=[var for var in df1numeric_data.columns if df1numeric_data[var].isna().sum()>0]
missing_null_var
```

#Distribution Plot

```python
plt.figure(figsize=(10,10))
sns.set()
for i,var in enumerate(missing_null_var):
  plt.subplot(2,2,i+1)
  sns.distplot(df1numeric_data[var],bins=20,kde_kws={'linewidth':5,'color':'#DC143C'})
```

After analysing above graph, we got to know that LotFrontage column is symmetic curve so we can take mean or median for fill null values.
####And other tow variables are skewed curve so we median is preferable there

```python
# first we fill all null values with mean
df2_mean=df1numeric_data.fillna(df1numeric_data.mean())
```

```python
df2_mean.isna().sum().sum()
```

###So all null values are filled 
#Now we compare we new dataset with original 

```python
plt.figure(figsize=(10,10))
sns.set()
for i,var in enumerate(missing_null_var):
  plt.subplot(2,2,i+1)
  sns.distplot(df1numeric_data[var],bins=20,kde_kws={'linewidth':5,'color':'red'},label="Original")
  sns.distplot(df2_mean[var],bins=20,kde_kws={'linewidth':3,'color':'green'},label='Mean')
  plt.legend()
```

#Now we fill with median

```python
df3_median=df1numeric_data.fillna(df1numeric_data.median())
```

```python
df3_median.isna().sum().sum()
```

```python
# now we plot new and original dataset
plt.figure(figsize=(10,10))
sns.set()
for i,var in enumerate(missing_null_var):
  plt.subplot(2,2,i+1)
  sns.distplot(df1numeric_data[var],bins=20,hist=False,kde_kws={'linewidth':8,'color':'red'},label="Original")
  sns.distplot(df3_median[var],bins=20,hist=False,kde_kws={'linewidth':5,'color':'black'},label='Median')
  sns.distplot(df2_mean[var],bins=20,hist=False,kde_kws={'linewidth':3,'color':'green'},label='Mean')
  plt.legend()
```

###In above graph we got to know that mean and median values are not varies
##They varies when we have some outliers

```python
# now we compare mean , median values with original 
df_concat=pd.concat([df1numeric_data[missing_null_var],df2_mean[missing_null_var],df3_median[missing_null_var]],axis=1)
```
