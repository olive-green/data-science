```python
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
```

```python
df
```

Now here dependent variable is  tip and all are independent variable. 
Now we are finding relationship b/w dependent and other independent variable with help of plotting.

```python
sns.relplot(x="total_bill",y="tip",data=df) 
#now here we are giving data=df b/c to tell 
```

```python
sns.relplot(x="total_bill",y="tip",data=df,hue="sex",style="time",size="size")
```

```python
sns.relplot(x="total_bill",y="tip",data=df,hue="smoker",col="time")
```

```python
sns.catplot(x="day",y="tip",data=df,kind="swarm")
```

```python
sns.catplot(x="day",y="tip",data=df,kind="swarm",hue="size")
```

```python
sns.catplot(x="total_bill",y="day",data=df,kind="violin",hue="sex")
```

```python
sns.catplot(x="total_bill",y="day",data=df,kind="violin",hue="sex",split=True)
```

```python
sns.catplot(x="total_bill",y="day",data=df,kind="violin",hue="sex",split=2)
```
