#Histogram

```python
population_ages=[22,35,40,33,32,31,45,21,22,26,24,28,34,36,37,33,38,32,34,24,22,23]
bins=[10,20,25,27,30,35,40,50,60,70] #bins are kind of batch or bucket like 10-20 mai kitne no. aa rhe hai here is zero, 20-25 mai kitne ar rhe hai here 7 coming
plt.hist(population_ages,bins,rwidth=0.8,color="r")
plt.show()
```

#PieChart

```python
slices=[7,5,14,12,19]
activities=["sleeping","writing","studying","playing","eating"]
colors=["g","r","c","m","b"]
plt.pie(slices,labels=activities,colors=colors)
plt.title("daily schedule")
plt.show()
```

```python
slices=[7,5,14,12,19]
activities=["sleeping","writing","studying","playing","eating"]
colors=["g","r","c","m","b"]
plt.pie(slices,labels=activities,colors=colors,shadow="true")
plt.title("daily schedule")
plt.show()
```

Explode take out the slice upward from rest of them

```python
slices=[7,5,14,12,19]
activities=["sleeping","writing","studying","playing","eating"]
colors=["g","r","c","m","b"]
plt.pie(slices,labels=activities,colors=colors,explode=(0,0.1,0,0,0))
plt.title("daily schedule")
plt.show()
```

 We use autopct to display the percent value using Python string formatting. For example, autopct='%1.1f%%' means that for each pie wedge, the format string is '1.1f%'. Try autopct='%1.2f%%' or autopct='%1.1f' to have a better idea about how it works.

```python
slices=[7,5,14,12,19]
activities=["sleeping","writing","studying","playing","eating"]
colors=["g","r","c","m","b"]
plt.pie(slices,labels=activities,colors=colors,explode=(0,0.1,0,0,0),autopct="%1.1f%%")
plt.title("daily schedule")
plt.show()
```

```python
slices=[7,5,14,12,19]
activities=["sleeping","writing","studying","playing","eating"]
colors=["g","r","c","m","b"]
plt.pie(slices,labels=activities,colors=colors,explode=(0,0.1,0,0,0),autopct="%1.2f%%")
plt.title("daily schedule")
plt.show()
```
