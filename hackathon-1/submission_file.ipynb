#Data Cleaning


#Null values percentage wise


```python
null_val_perc=df.isna().sum()/df.shape[0]*100
null_val_perc
```

```python
# delete null columns having perc > 62
deleted_col=null_val_perc[null_val_perc > 62].keys()
deleted_col
```

```python
df=df.drop(columns=deleted_col)
```

```python
df.shape
```

```python
df.head()
```

```python
# numeric var
num_var=df.select_dtypes(include=["int64","float64"])
num_var.head()
```

Now we plot heatmap to check null values in num var

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(num_var.isna())
```

#filling null values

```python
# take out numeric null columns
num_null_var=[var for var in num_var.columns if num_var[var].isna().sum()>0]
num_null_var
```

```python
for var in num_null_var:
  df[var].fillna(df[var].median(),inplace=True)
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(df.isna())
```

#now filling categorical variables

```python
cat_var=df.select_dtypes(include=["object"])
cat_var
```

```python
cat_var.shape
```

```python
null_perc=cat_var.isna().mean()*100
cat_null_var=null_perc[null_perc>0].keys()
cat_null_var
```

```python
cat_null_var.shape
```

```python
# fill all null columns
for var in cat_null_var:
  df[var].fillna(df[var].mode()[0],inplace=True)
```
