#Data Cleaning


#Null values percentage wise


```python
null_val_perc=df.isna().sum()/df.shape[0]*100
null_val_perc
```

```python
# delete null columns having perc > 62
deleted_col=null_val_perc[null_val_perc > 62].keys()
deleted_col
```

```python
df=df.drop(columns=deleted_col)
```

```python
df.shape
```

```python
df.head()
```

```python
# numeric var
num_var=df.select_dtypes(include=["int64","float64"])
num_var.head()
```

Now we plot heatmap to check null values in num var

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(num_var.isna())
```

#filling null values

```python
# take out numeric null columns
num_null_var=[var for var in num_var.columns if num_var[var].isna().sum()>0]
num_null_var
```

```python
for var in num_null_var:
  df[var].fillna(df[var].median(),inplace=True)
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(df.isna())
```

#now filling categorical variables

```python
cat_var=df.select_dtypes(include=["object"])
cat_var
```

```python
cat_var.shape
```

```python
null_perc=cat_var.isna().mean()*100
cat_null_var=null_perc[null_perc>0].keys()
cat_null_var
```

```python
cat_null_var.shape
```

```python
# fill all null columns
for var in cat_null_var:
  df[var].fillna(df[var].mode()[0],inplace=True)
  print("filling",var,"with value",df[var].mode()[0])
```

#Train the model

```python
# Assign X as a DataFrame of features and y as a Series of the outcome variable
X=df.drop("LICENSE STATUS",axis=1)
y=df["LICENSE STATUS"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=1)
```

#Feature scaling

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
# y_train=sc.fit_transform(y_train)
```

```python
# from sklearn.linear_model import LogisticRegression
# classifier=LogisticRegression()
# classifier.fit(X_train,y_train)

from sklearn.tree import DecisionTreeClassifier
classifier=DecisionTreeClassifier()
classifier.fit(X_train,y_train)
```

```python
y_pred=classifier.predict(X_test)
```

```python
classifier.score(X_test,y_test)
```
